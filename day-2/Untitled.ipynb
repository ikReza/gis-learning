```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("default")
```

```python
region_of_interest_1 = gpd.read_file("Shapefiles/region_of_interest_1.shp")
```

```python
region_of_interest_1
```

```python
region_of_interest_1.plot(edgecolor="black", cmap="ocean");
```

```python
region_of_interest_2 = gpd.read_file("Shapefiles/region_of_interest_2.shp")
```

```python
fig, ax = plt.subplots()
region_of_interest_1.plot(ax=ax, edgecolor="black", cmap="ocean")
region_of_interest_2.plot(ax=ax, edgecolor="black", cmap="summer", alpha=0.6);
```

```python
region_of_interest_1.geometry, region_of_interest_2.geometry
```

We can see that both of the shape is polygon.
